```python
from matplotlib import pyplot as plt
#import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[10,20,30]
y=[10,20,30]
plt.plot(x,y)
plt.title("sample chart")
plt.show()
```

```python
x=[10,20,30]
y=[10,20,30]
plt.plot(x,y)
plt.title("sample chart")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
#plt.xticks(x,["a","b","c"])
#plt.yticks(y,[100,200,300])
plt.xlim([10,100])
plt.ylim([1000,2000])
plt.show()
```

```python
x = [1,2,3]
y1= ["a","b","c"]
y2= ["x","y","z"]
y3= [10,20,30]
y4= [100,200,300]
plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)

plt.show()
```

```python
x = [1,2,3]
y= ["a","b","c"]
plt.plot(x,y)
f1={'family':'cambria','color':'red','size':20}
f2={'family':'calibri','color':'blue','size':15}
f3={'family':'calibri','color':'green','size':15}
plt.title("sample chart",fontdict=f1)
plt.xlabel("X-Axis",fontdict=f2)
plt.ylabel("Y-Axis",fontdict=f3)
plt.show()
```

```python
x=[10,20,30]
y=[10,20,30]
plt.bar(x,y)
plt.title("sample chart")
plt.show()
```
